# Carga Condicional de Librerias

```python
require_pkg <- function(pkg) {
  if (!require(pkg, character.only = TRUE)) {
    install.packages(pkg, dependencies = TRUE)
    library(pkg, character.only = TRUE)
  } else {
    library(pkg, character.only = TRUE)
    message(paste(pkg, "ya está instalado."))
  }
}

# Librerías base
require_pkg("tidyverse")
require_pkg("GGally")
require_pkg("factoextra")
require_pkg("survival")
require_pkg("survminer")
```

# Dataset

```python
# Dataset: Heart Failure Clinical Data
url_dataset <- "https://raw.githubusercontent.com/ageron/data/main/heart_failure_clinical_records_dataset.csv"

dataset <- read.csv(url_dataset)

# Exploración inicial
cat("DATASET HEART FAILURE\n")
cat("Dimensiones:\n")
print(dim(dataset))

cat("\nColumnas:\n")
print(colnames(dataset))

cat("\nTipos de datos:\n")
print(str(dataset))

cat("\nValores faltantes por columna:\n")
print(colSums(is.na(dataset)))

cat("\nPrimeras filas:\n")
head(dataset)

cat("\nÚltimas filas:\n")
tail(dataset)

cat("\nResumen estadístico:\n")
summary(dataset)
```
